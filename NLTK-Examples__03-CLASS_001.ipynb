```python
s1 = "Do yourself a favor and Stay as far away from OPPO as you can. It plays and sounds great while it runs but will inevitably stop working after a few months. I should have listened to my guts and bought a brand name product like Sony or Panasonic. Now I'm stuck with a player that reguses to work and a company that is shut down (probably because they realised they sold a crappy product and don't want to pay yo fix thousands of in-op players). Guess that's the chinese way after all"
```

```python
s2 = "I, too, hesitated at the price, but I wanted something that would play everything I had and this fit the bill. It does a great job and I have had no troubles with it at all, now a year in. My only (minor) complaint is that all the buttons on the remote light up every time you move the remote. That tends to wear on the batteries if you are fast forwarding or otherwise using the remote for long periods. I also use rechargeable batteries since the remote is the heaviest of any of my others. All in all, this is a great player, easy to understand installation, and performs exceedingly well"
```

```python
print(s1) # bad review
```

```python
print(s2) # good review
```

```python
####################
# how is the total sentiment changes when using r/v/n/a/none
s = s1

from nltk.corpus import stopwords

pos = 0.0
neg = 0.0
total = 0.0
words = [word for word in s.split()]# if word.lower() not in stopwords.words('english')]
for wr in words:
    ls = list(swn.senti_synsets(wr,'r')) # adverb (try with r/v/n/a)
    for s in ls:
        r = swn.senti_synset(s.synset.name())
        pos += r.pos_score()
        neg += r.neg_score()
        
print("pos =",pos,"neg =",neg)   
```

```python
import nltk.corpus as nc
# https://wordnet.princeton.edu
car_synsets = nc.wordnet.synsets('car')
print (car_synsets)
```

```python
dog = nc.wordnet.synset('dog.n.01')
cat = nc.wordnet.synset('cat.n.01')

similarity_score1 = dog.path_similarity(cat) # path distance similarity
similarity_score2 = dog.wup_similarity(cat) # Wu - Palmer similarity
print(similarity_score1)
print(similarity_score2)
```

```python
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures

words =['the','quick','brown','fox',
'jumps','over','the','lazy','dog']

bigram_finder = BigramCollocationFinder.from_words(words)
bigrams = bigram_finder.nbest(BigramAssocMeasures.chi_sq, 3)
bigrams
```

```python
text = "tramp says that tramp says that the war is happening"
finder = BigramCollocationFinder.from_words(nltk.word_tokenize(text))
```

```python
for k,v in finder.ngram_fd.items():
    print(k[0],k[1],v)
```
